# Append start latitude and longitude for 2016 using 2021 as reference

```python
x = df_2021

#x['start_station_id'] = pd.to_numeric(x['start_station_id'])
x = x.dropna()
x['start_station_id'] = x['start_station_id'].astype(int)

x = x.drop_duplicates(subset=['start_station_id'])

x = x[['start_station_id' , 'start_lat', 'start_lng']]

x = x.rename(columns = {'start_station_id':'Start.station.number'})

y = df_2016.merge(x , on='Start.station.number', how='left')
y
```

# Append end latitude and longitude for 2016 using 2021 as reference

```python
x = df_2021

#x['start_station_id'] = pd.to_numeric(x['start_station_id'])
x = x.dropna()
x['end_station_id'] = x['end_station_id'].astype(int)

x = x.drop_duplicates(subset=['end_station_id'])

x = x[['end_station_id' , 'end_lat', 'end_lng']]

x = x.rename(columns = {'end_station_id':'End.station.number'})

y = y.merge(x , on='End.station.number', how='left')
y
```

```python
y["start_lng"].isna().sum()
y["end_lng"].isna().sum()
```

```python
y.to_csv("../data/2016/df_2016_geospatial.csv", index=False) # about 5.5k NA's for both start and end geo coords
```
