```python
import pandas as pd

# Before combining the new dataframe (songs post 2015) with the old dataframe, we should make sure everything is clean in both datasets

# Start by cleaning the new dataset since the old one has been probably cleaned by the GitHub users


# Note: This notebook will focus on cleaning text data (lyrics of songs). The remaninig parts of cleaning are done in a seperate R script 

new_data = pd.read_csv('Data/final_df.csv')


# Keep only the following columns: lyrics (for analysis), year, and title

lyric_data = new_data[['title' , 'lyrics' , 'year']]



```

```python
# Clean the lyrics of songs (remove stop words, etc.)

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

sw_nltk = stopwords.words('english')
print(sw_nltk)


for i in range(len(lyric_data)):

    query = lyric_data['lyrics'][i].replace("\n" , " ")

    querywords = query.split()  

    resultwords  = [word for word in querywords if word.lower() not in sw_nltk]

    result = ' '.join(resultwords)

    lyric_data['lyrics'][i] = result
```

```python
# Count vectorization

# The analysis will be done per decade, so it only makes sense to split the lyrics dataframe per decade 

from sklearn.feature_extraction.text import CountVectorizer

lyric_data_50s =lyric_data[(lyric_data['year'] > 1950) & (lyric_data['year'] <1960)].reset_index()
lyric_data_60s = lyric_data[(lyric_data['year'] > 1960) & (lyric_data['year'] <1970)].reset_index()
lyric_data_70s = lyric_data[(lyric_data['year'] > 1970) & (lyric_data['year'] <1980)].reset_index()
lyric_data_80s = lyric_data[(lyric_data['year'] > 1980) & (lyric_data['year'] <1990)].reset_index()
lyric_data_90s = lyric_data[(lyric_data['year'] > 1990) & (lyric_data['year'] <2000)].reset_index()
lyric_data_00s = lyric_data[(lyric_data['year'] > 2000) & (lyric_data['year'] <2010)].reset_index()
lyric_data_2010s = lyric_data[(lyric_data['year'] > 2010)].reset_index()



# INITIALIZE COUNT VECTORIZER for 50s music

vectorizer=CountVectorizer() 
matrix  =  vectorizer.fit_transform([lyric_data_50s['lyrics'][0]]) 

counts = pd.DataFrame(matrix.toarray(),
                      columns=vectorizer.get_feature_names())


x = pd.DataFrame(counts.T)


for i in range(len(lyric_data_50s['lyrics'])):

    vectorizer=CountVectorizer() 
    matrix  =  vectorizer.fit_transform([lyric_data_50s['lyrics'][i]]) 

    counts = pd.DataFrame(matrix.toarray(),
                        columns=vectorizer.get_feature_names())


    y = pd.DataFrame(counts.T)

    x = x.append(y)

   




```
