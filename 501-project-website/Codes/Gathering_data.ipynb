```python
class SpotifyAPI(object):
    access_token = None
    access_token_expires = datetime.datetime.now()
    access_token_did_expire = True
    client_id = None
    client_secret = None
    token_url = "https://accounts.spotify.com/api/token"
    
    def __init__(self, client_id, client_secret, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.client_id = client_id
        self.client_secret = client_secret

    def get_client_credentials(self):
        """
        Returns a base64 encoded string
        """
        client_id = self.client_id
        client_secret = self.client_secret
        if client_secret == None or client_id == None:
            raise Exception("You must set client_id and client_secret")
        client_creds = f"{client_id}:{client_secret}"
        client_creds_b64 = base64.b64encode(client_creds.encode())
        return client_creds_b64.decode()
    
    def get_token_headers(self):
        client_creds_b64 = self.get_client_credentials()
        return {
            "Authorization": f"Basic {client_creds_b64}"
        }
    
    def get_token_data(self):
        return {
            "grant_type": "client_credentials"
        } 
    
    def perform_auth(self):
        token_url = self.token_url
        token_data = self.get_token_data()
        token_headers = self.get_token_headers()
        r = requests.post(token_url, data=token_data, headers=token_headers)
        if r.status_code not in range(200, 299):
            return False
        data = r.json()
        now = datetime.datetime.now()
        access_token = data['access_token']
        expires_in = data['expires_in'] # seconds
        expires = now + datetime.timedelta(seconds=expires_in)
        self.access_token = access_token
        self.access_token_expires = expires
        self.access_token_did_expire = expires < now
        return True
        
```

# Part Four: Gathering additional song features

```python
# Import the updated dataframe that has the additional features from R
import pandas as pd
df = pd.read_csv('/Users/anthonymoubarak/Desktop/Project/Project Data/missingyears_updated.csv')
```

```python
# There still are a couple of features that need to be added using python in order to have the total number
# of features as all the songs we have gathered previously (pre 2015)

# We will start with basic setiment analysis on the lyrics to give us the 4 parameters:
# neg,neu,pos,compound

import nltk
import nltk.sentiment.vader


# Create the sent analyzer object 
sent_analyzer = nltk.sentiment.vader.SentimentIntensityAnalyzer()
```

```python
# Create 4 new columns in our dataframe to store the results from the analyzer

df['pos'] = ''
df['neg'] = ''
df['neu'] = ''
df['compound'] = ''
```

```python
for i in range(len(df['lyrics'])):
    
    try: #Some songs do not have lyrics. Until this is resolved,  try has to be included
        result = sent_analyzer.polarity_scores(df['lyrics'][i])
        pos = result['pos']
        neu = result['neu']                                       
        neg = result['neg']
        comp = result['compound']  
        df['pos'][i] = pos
        df['neg'][i] = neg
        df['neu'][i] = neu
        df['compound'][i] = comp
    except:
        continue
```

```python
# The next step is gathering readability features using the textstat package 

from textstat.textstat import textstat as ts
```

```python
# Create empty columns in the dataframe for the to be added features, which are:
# num_words , num_lines, num_syllables, difficult_words,fog_index, flesch_index ,f_k_grade

df['num_words'] = ''
df['num_lines'] = ''
df['num_syllables'] = ''
df['difficult_words'] = ''
df['fog_index'] = ''
df['flesch_index'] = ''
df['f_k_grade'] = ''


for i in range(len(df['lyrics'])):
    
    try: #Some songs do not have lyrics. Until this is resolved,  try has to be included
        fk = ts.flesch_kincaid_grade(df['lyrics'][i])
        fl = ts.flesch_reading_ease(df['lyrics'][i])
        fog = ts.gunning_fog(df['lyrics'][i])
        diff = ts.difficult_words(df['lyrics'][i])
        syll = ts.syllable_count(df['lyrics'][i])
        words = ts.lexicon_count(df['lyrics'][i], True)
        lines = ts.sentence_count(df['lyrics'][i])
        
        df['num_words'][i] = fk
        df['num_lines'][i] = fl
        df['num_syllables'][i] = syll
        df['difficult_words'][i] = diff
        df['fog_index'][i] = fog
        df['flesch_index'][i] = words
        df['f_k_grade'][i]= fk
        
     
    except:
        continue
```
